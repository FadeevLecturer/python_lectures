## Линейные графики

Построение кривых --- один из самых часто встречающихся видов графиков. За построение таких графиков `Matplotlib` отвечает метод [plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html). Этот метод строит кривую из заданного набора точек, соединяя их между собой линией, т.е. чтобы построить, например, график функции $f(x)$, необходимо в первую очередь значения этой функции табулировать. 

На вход метод `plot` ожидает два массива чисел: первый массив отвечает за $x$-координаты точек кривой, а второй за $y$-координаты. Кроме того можно сразу указать опциональными параметрами внешний вид кривой: её цвет, толщину, стиль, наличие маркеров, их форму, размер и многое другое. 

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.sin(2*x)

fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(x, y1, "b", label=r"$\sin(x)$")
ax.plot(x, y2, "g--", label=r"$\sin(2x)$")
ax.legend()
```

# Основные виды графиков

Существует огромное разнообразие видов графиков. Оптимальный вид графика может зависеть и от природы данных и от того, какую идею этим графиком вы хотите проиллюстрировать. Количество готовых методов в `Matplotlib` велико и запомнить их все не выйдет. На выручку приходит [галерея примеров](https://matplotlib.org/stable/gallery/index.html) из официальной документации, в которой нередко можно найти пример создания необходимого графика, который можно без особого труда применить для своих данных и при необходимости доработать детали.

## Кривые

Построение кривых --- один из самых часто встречающихся видов графиков. За построение таких графиков `Matplotlib` отвечает метод [plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html). В качестве аргументов этот метод принимает два одномерных массива одинаковой длинны $x$ и $y$, элементы которых представляют координаты $x_i$ и $y_i$ точек кривой. Таким образом, если требуется построить график функции $y=f(x)$, то необходимо сначала табулировать значения этой функции.

В качестве примера построим график функции $y=\sin(e^x)$ на отрезке от 0 до $\pi$. Для этого разобьём этот отрезок на 30 отрезков методом [numpy.linspace](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html). Вычислить значение функции $\sin(e^x)$ в узлах этой сетки можно с помощью векторизации. 

```python
import numpy as np
from matplotlib import pyplot as plt

n_points = 30
x = np.linspace(0, np.pi, n_points)
y = np.sin(np.exp(x))

fig, ax = plt.subplots(layout="tight")
ax.plot(x, y, label="y=$\sin(e^x)$")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_title(f"График по {n_points} точкам")
ax.legend()
```

Видим, что график получился угловатым, так как `Matplotlib` просто соединяет переданные точки отрезками. Чтобы сгладить график, можно просто изначально задать более густую сетку. Увеличим количество узлов в сетке до 300.

```python
import numpy as np
from matplotlib import pyplot as plt

n_points = 300
x = np.linspace(0, 1*np.pi, n_points)
y = np.sin(np.exp(x))

fig, ax = plt.subplots(layout="tight")
ax.plot(x, y, label="y=$\sin(e^x)$")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_title(f"График по {n_points} точкам")
ax.legend()
```

За построение кривых в пространстве отвечает тот же самый метод, но у трехмерных осей. В качестве примера построим график параметрически заданной на отрезке $t\in[0, 2\pi]$ кривой

$$
\begin{cases}
x(t) = \sin(3t)\cos(5t), \\
y(t) = \sin(3t)\sin(5t), \\
z(t) = \cos(3t).
\end{cases}
$$

```python
import numpy as np
from matplotlib import pyplot as plt

n_points = 300
t = np.linspace(0, 2*np.pi, n_points)
x = np.sin(3*t)*np.cos(5*t)
y = np.sin(3*t)*np.sin(5*t)
z = np.cos(3*t)

fig, ax = plt.subplots(figsize=(7, 7), layout="tight", subplot_kw={"projection": "3d"})
ax.plot(x, y, z, label="y=$\sin(e^x)$")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$z$")
ax.set_title(f"График по {n_points} точкам")
ax.legend()
```

Стиль кривой удобнее указывать опциональными параметрами метода `plot`.
